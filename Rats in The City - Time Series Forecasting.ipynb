```python
import os 
import pandas as pd
import numpy as np
```

Load Data

```python
df = pd.read_pickle("./Rat_Sightings.pkl")
print(df.shape)
df.head()
```

```python
# want to do analysis on each borough
time_data = (df
             .assign(n=0)
             .groupby(["Borough", "Year", "Month"])
             .n
             .count()
             .reset_index()
             .sort_values(["Year", "Borough"])
             .reset_index(drop=True)
             .rename(columns={"n":"Count"})
            )

time_data["Date"] = time_data.Month+" "+time_data.Year
time_data.Date = pd.to_datetime(time_data.Date)

print(time_data.shape)
time_data.head()
```

Time Series Plot

```python
import plotly.plotly as py
import plotly.graph_objs as go

# subset data by boroughs
bronx_data = time_data.query("Borough == 'Bronx'").sort_values("Date")
bk_data = time_data.query("Borough == 'Brooklyn'").sort_values("Date")
mnhtn_data = time_data.query("Borough == 'Manhattan'").sort_values("Date")
qns_data = time_data.query("Borough == 'Queens'").sort_values("Date")
si_data = time_data.query("Borough == 'Staten Island'").sort_values("Date")

trace_bronx = go.Scatter(
    x=bronx_data.Date,
    #pd.to_datetime
    y=bronx_data.Count,
    name="Bronx",
    line = dict(color = "grey"),
    opacity=0.8
)

trace_brooklyn = go.Scatter(
    x=bk_data.Date,
    #pd.to_datetime
    y=bk_data.Count,
    name="Brooklyn",
    line = dict(color = "navy"),
    opacity=0.8
)

trace_manhattan = go.Scatter(
    x=mnhtn_data.Date,
    #pd.to_datetime
    y=mnhtn_data.Count,
    name="Manhattan",
    line = dict(color = "silver"),
    opacity=0.8
)

trace_queens = go.Scatter(
    x=qns_data.Date,
    #pd.to_datetime
    y=qns_data.Count,
    name="Queens",
    line = dict(color = "green"),
    opacity=0.8
)

trace_staten_isl = go.Scatter(
    x=si_data.Date,
    #pd.to_datetime
    y=si_data.Count,
    name="Staten Island",
    line = dict(color = "black"),
    opacity=0.8
)

data = [trace_bronx, trace_brooklyn, trace_manhattan,
        trace_queens, trace_staten_isl
       ]

layout = dict(
    title="NYC Time Series of Rat Data",
    xaxis=dict(
        rangeselector=dict(
            buttons=list([
                dict(count=1,
                     label="1 month",
                     step="month",
                     stepmode="backward"
                    ),
                dict(count=3,
                     label = "3 months",
                     stepmode="backward"
                    ),
                dict(count=6,
                     label="6 months",
                     stepmode="backward"
                    ),
                dict(count=12,
                     label="Year",
                     stepmode="backward"
                    ),
                dict(step="all")
            ])
        ),
        rangeslider=dict(),
        type="date"
    )
)

fig = dict(data=data, layout=layout)
py.iplot(fig)
```

```python
"""# facebooks prophet package is a great time series package for 
analyzing time series data with daily observations that can exhibit 
different patters of varying scales. 
"""
import fbprophet

# group data by borough/date
rat_data = (df
            .assign(n=0)
            .rename(columns={"Created Date":"ds"})
            .groupby(["Borough","ds"])
            .n
            .count()
            .reset_index()
            .rename(columns={"n":"y"})
            )

# change type 
rat_data.y = rat_data.y.astype(float)

# view
rat_data.head()
```

```python
# subset the data for the boroughs
bronx_data = rat_data.query("Borough == 'Bronx'").sort_values("ds")
bk_data = rat_data.query("Borough == 'Brooklyn'").sort_values("ds")
mnhtn_data = rat_data.query("Borough == 'Manhattan'").sort_values("ds")
qns_data = rat_data.query("Borough == 'Queens'").sort_values("ds")
si_data = rat_data.query("Borough == 'Staten Island'").sort_values("ds")

# instantiate the models
bronx_rat_prophet = fbprophet.Prophet()
bk_rat_prophet = fbprophet.Prophet()
mnhtn_rat_prophet = fbprophet.Prophet()
qns_rat_prophet = fbprophet.Prophet()
si_rat_prophet = fbprophet.Prophet()

# fit the models
bronx_rat_prophet.fit(bronx_data)
bk_rat_prophet.fit(bk_data)
mnhtn_rat_prophet.fit(mnhtn_data)
qns_rat_prophet.fit(qns_data)
si_rat_prophet.fit(si_data)
```

```python
# make dataframe of future preds
bronx_rat_forecast = bronx_rat_prophet.make_future_dataframe(periods=365*2, freq="D")
bk_rat_forecast = bk_rat_prophet.make_future_dataframe(periods=365*2, freq="D")
mnhtn_rat_forecast = mnhtn_rat_prophet.make_future_dataframe(periods=365*2, freq="D")
qns_rat_forecast = qns_rat_prophet.make_future_dataframe(periods=365*2, freq="D")
si_rat_forecast = si_rat_prophet.make_future_dataframe(periods=365*2, freq="D")

# make predictions
bronx_rat_forecast = bronx_rat_prophet.predict(bronx_rat_forecast)
bk_rat_forecast = bk_rat_prophet.predict(bk_rat_forecast)
mnhtn_rat_forecast = mnhtn_rat_prophet.predict(mnhtn_rat_forecast)
qns_rat_forecast = qns_rat_prophet.predict(qns_rat_forecast)
si_rat_forecast = si_rat_prophet.predict(si_rat_forecast)
```
